### t검정 

- 집단간 평균비교

### compact 자동차와 suv자동차의 도시연비가 통계적으로 유의한 관계를 가지고 있는지

```python
import pandas as pd
mpg = pd.read_csv('mpg.csv')
```

```python
mpg.query('category in ["compact", "suv"]') \
   .groupby('category', as_index = False) \
   .agg(n    = ('category', 'count'),
        mean = ('cty', 'mean'))
```

- compact와 suv 피처를 query검색하여 t검정 변수로 설정

```python
compact = mpg.query('category == "compact"')['cty']
suv = mpg.query('category == "suv"')['cty']
```

```python
from scipy import stats
stats.ttest_ind(compact, suv, equal_var = True)
```

**p-valuue 유의확률이 0.05이상이므로 신뢰수준밖의 통계결과=즉 통계적으로 유의하지않고, 우연일확률 높다고 판단할수있다**

### 일반 휘발유와 고급 휘발유의 도시 연비 t 검정

- 일반휘발유 fl피처의 r변수,  고급휘발유 fl피처의 p변수

```python
mpg.query('fl in ["r", "p"]') \
   .groupby('fl', as_index = False) \
   .agg(n    = ('category', 'count'),
        mean = ('cty', 'mean'))
```

```python
regular = mpg.query('fl == "r"')['cty']
premium = mpg.query('fl == "p"')['cty']
type(regular)
```

```python
# t-test
stats.ttest_ind(regular, premium, equal_var = True)
```

- 변수간 통계적으로 유의한 상관관계가 얼마정도 되는지 수치로 나타내며, 연속형 변수간에 측정가능하다

### 실업자 수와 개인소비 지출의 상관관계

```python
# economics 데이터 불러오기
economics = pd.read_csv('economics.csv')

# 상관행렬 만들기
economics[['unemploy', 'pce']].corr()
```

### 상관관계를 히트맵으로 시각화 하기

```python
mtcars = pd.read_csv('../Data/mtcars.csv')
mtcars.head()
```

- 1. 상관행렬 만들기

```python
car_cor = mtcars.corr()      # 상관행렬 만들기
car_cor = round(car_cor, 2)  # 소수점 둘째 자리까지 반올림
car_cor
```

- 2.히트맵 객체 생성

```python
import matplotlib.pyplot as plt
plt.rcParams.update({'figure.dpi' : '120',           # 해상도 설정
                     'figure.figsize': [7.5, 5.5]})  # 가로 세로 크기 설정
```

```python
# 히트맵 만들기
import seaborn as sns
sns.heatmap(car_cor,
            annot = True,   # 상관계수 표시
            cmap = 'RdBu')  # 컬러맵
```

**행렬 특성상 히트맵 위대각행렬에는 상관값이 n:n = 1이거나 아래 대각행렬과 겹치는 값이 있으므로 위대각행렬을 제거해야한다**

- 3.mask객체를 이용해 히트맵에 위대각행렬 제거

```python
# mask 만들기
import numpy as np
mask = np.zeros_like(car_cor)
mask
```

```python
# 오른쪽 위 대각 행렬을 1로 바꾸기
mask[np.triu_indices_from(mask)] = 1
mask
```

- 4.히트맵에 마스크 행렬 적용

**히트맵은 1을 mask로 적용**

```python
# 히트맵 만들기
sns.heatmap(data = car_cor,
            annot = True,   # 상관계수 표시
            cmap = 'RdBu',  # 컬러맵
            mask = mask)    # mask 적용
```

- 5.마스크때문에 지워진 첫행, 마지막 열 처리하기

```python
mask_new = mask[1:, :-1]         # mask 첫 번째 행, 마지막 열 제거
cor_new = car_cor.iloc[1:, :-1]  # 상관행렬 첫 번째 행, 마지막 열 제거

# 히트맵 만들기
sns.heatmap(data = cor_new,
            annot = True,       # 상관계수 표시
            cmap = 'RdBu',      # 컬러맵
            mask = mask_new)    # mask 적용
```

- 6.범례 크기 줄이기,경계구분선추가

```python
sns.heatmap(data = cor_new,
            annot = True,       # 상관계수 표시
            cmap = 'RdBu',      # 컬러맵
            mask = mask_new,
            linewidths = .5, #경계구분선
            cbar_kws = {'shrink': .5})    # mask 적용
```
