### 1. 02_Data.csv파일을 불러와, df1으로 선언하고 데이터의 구조와 타입을 확인하시오.

```python
import pandas as pd
df1 = pd.read_csv('02_Data.csv')

df1.info()
```

### 2. '프로그램가격'의 평균, 최대값, 최소값을 확인하시오.

```python
df1['프로그램가격'].describe()
```

### 3. '채널'의 항목과 빈도수를 확인하고, '채널'별 '최종결제금액'의 합을 계산하여, result1.xlsx로 저장하시오.

```python
df1['채널'].value_counts()
```

```python
df2 = df1.pivot_table(index = '채널', values = '최종결제금액', aggfunc = 'sum')
df2 = df2.sort_values(by='최종결제금액', ascending = False).reset_index()

df2.to_excel('result1.xlsx')
df2
```

### 4. '채널'의 빈도수가 높은 상위 3개 항목을 뽑아, 해당 데이터의 '보조식품(추가구매)'에 따른 '최종결제금액'의 합과 평균을 계산하시오.  

```python
tc = df1['채널'].value_counts().head(3).index

threec = df1[df1['채널'].isin(tc)]
result = threec.groupby('보조식품(추가구매)')['최종결제금액'].agg(['sum', 'mean'])

print(result)
```

### 5. '결제시각'데이터에서 날짜와 시간데이터를 나누어 데이터만 뽑아, '결제날짜'와 '결제시간'파생변수로 각각 선언하시오.

```python
Date = df1['결제시각']
df1['결제날짜'] = df1['결제시각'].str.slice(start=0, stop=10)
df1['결제시간'] = df1['결제시각'].str.slice(start=10)

df1
```

### 6. 앞서 추출된 '결제날짜'데이터를 이용하여, '연도' / '월' / '일' 데이터를 추출하시오.

```python
df1['결제날짜'] = pd.to_datetime(df1['결제날짜'])

df1['Year'] = df1['결제날짜'].dt.year
df1['Month'] = df1['결제날짜'].dt.month
df1['Day'] = df1['결제날짜'].dt.day

df1
```

### 7. 결제 '월'별 '최종결제금액'의 합을 계산하시오.

```python
df1.pivot_table(index = 'Month', values = '최종결제금액', aggfunc = 'sum').reset_index()
```

### 8. '최종결제금액'이 높은 30명의 고객 명단을 추출하여, result2.xlsx로 선언하시오.

```python
df8 = df1.sort_values(by='최종결제금액', ascending = False).head(30)
df8.to_excel('result2.xlsx')
df8
```

### 9. '상태'가 completed에 해당하는 데이터만 뽑아, '채널'별 '최종결제금액'의 합을 계산하시오.

```python
df9 = df1.query('상태 == "completed"')
df9.pivot_table(index = '채널', values = '최종결제금액', aggfunc = 'sum').reset_index()
```

```python
df1.query('상태 == "completed"')\
    .groupby('채널')\
    .agg(sum = ('최종결제금액', 'sum'))\
    .sort_values('sum', ascending = False)\
    .reset_index()
```
