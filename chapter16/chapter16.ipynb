<h2>16-1 [ ]이용하기</h2>

```python
import pandas as pd
df_raw = pd.read_csv('D:/WE_IT/doit/bookdata/exam.csv')
df = df_raw.head(10)
df
```

<h3>조건을 충족하는 행 추출하기</h3>

```python
# nclass 추출
df['nclass']
```

```python
# nclass가 1인지 확인
df['nclass'] == 1
```

```python
# nclass가 1이면 추출
df[df['nclass'] == 1]
```

```python
# 수학 점수가 80점 이상이면 추출
df[df['math'] >= 80]
```

```python
# nclass가 1이면서 수학 점수가 50점 이상
df[(df['nclass'] == 1) & (df['math'] >= 50)]
```

```python
# 수학 점수가 50점 미만이거나 영어 점수가 50점 미만
df[(df['math'] < 50) | (df['english'] < 50)]
```

```python
df['nclass']
```

```python
# 여러 변수를 동시에 추출
df[['id', 'nclass']]
```

```python
# 시리즈로 추출
df['id']
```

```python
# 데이터 프레임으로 추출
df[['id']]
```

<h3>조건을 충족하는 행에서 열 추출하기</h3>

```python
# nclass가 1인 행의 math열
df[df['nclass'] == 1]['math']
```

```python
# nclass가 1인 행의 math, english열
df[df['nclass'] == 1][['math', 'english']]
```

```python
# pandas 함수만 사용
df.groupby('nclass') \
  .agg(math    = ('math', 'mean'),
       english = ('english', 'mean'))
```

```python
# pandas 함수와 [] 함께 사용
df.groupby('nclass')[['math', 'english']].mean()
```

<h2>16-2 df.loc[ ]이용하기</h2>

<h3>인덱스 활용하기</h3>

```python
# 인덱스 문자열 지정하기
df = pd.DataFrame({'var1' : [1, 2, 3],
                   'var2' : [4, 5, 6]},
                   index = ['kim', 'lee', 'park'])
df
```

<h3>인덱스 문자열을 지정해 행 추출하기</h3>

```python
# kim행 추출
df.loc['kim']
```

```python
# 여러행(kim, park행) 추출
df.loc[['kim', 'park']]
```

<h3>인덱스 문자열을 지정해 열 추출하기</h3>

```python
# lee행의 var1열 추출
df.loc['lee', 'var1']
```

```python
# kim, lee행의 var2열 추출
df.loc[['kim', 'lee'], 'var2']
```

```python
df.loc[,'var1']    # 에러 발생
```

```python
# 모든 행은 :을 입력
df.loc[:, 'var1']
```

<h3>인덱스 번호를 지정해 행 추출하기</h3>

```python
# 인덱스 번호가 있는 데이터 프레임 (df_raw는 383쪽의 데이터 프레임)
df = df_raw.copy()
df
```

```python
# 0행 추출
df.loc[0]
```

```python
# 2행 추출
df.loc[2]
```

```python
# 시리즈로 추출
df.loc[0]
```

```python
# 데이터 프레임으로 추출
df.loc[[0]]
```

```python
# 여러 행(0, 3, 5행 추출)
df.loc[[0, 3, 5]]
```

```python
# 0~3행 추출
df.loc[0:3]
```

```python
# 7~9행 추출
df.loc[7:9]
```

```python
# : 앞에 아무 값도 입력하지 않으면 첫 번째 행부터 출력됨
df.loc[:3]  # df.loc[0:3]과 동일
```

```python
# : 뒤에 아무 값도 입력하지 않으면 마지막 행까지 추출함
df.loc[5:] 
```

```python
# : 앞뒤에 아무 값도 입력하지 않으면 첫 번째 행부터 마지막 행까지 모두 추출함
df.loc[:]  # df와 출력 결과가 동일함
```

```python
# df.loc[]는 인덱스 번호로 행은 추출할 수 있지만, 열은 추출할 수 없음
df.loc[2,:]  # 행 추출 가능
```

```python
# df.loc[]는 인덱스 번호로 행은 추출할 수 있지만, 열은 추출할 수 없음
df.loc[:,2]  # KeyError 발생
```

<h3>조건을 충족하는 행 추출하기</h3>

```python
# nclass가 1이면 추출
df.loc[df['nclass'] == 1]
```

```python
# 수학 점수가 60점 이상이면 추출
df.loc[df['math'] >= 60]
```

```python
# nclass가 1인 행의 math, english열 추출
df.loc[df['nclass'] == 1, ['math', 'english']]
```

```python
# 0~3행의 math열 추출
df.loc[0:3, 'math']
```

<h4>조건을 충족하는 행 추출 방법 비교</h4>
수학 점수가 50을 초과하는 행 추출하기

```python
df.query('math > 50')    # df.query()
```

```python
df[df['math'] > 50]      # df
```

```python
df.loc[df['math'] > 50]  # df.loc[]
```

```python
# 인덱스 문자열이 있는 데이터 프레임
df_label = pd.DataFrame({'var1' : [1, 2, 3],
                         'var2' : [4, 5, 6]},
                         index = ['kim', 'lee', 'park'])
df_label
```

```python
# 2행의 3열 추출
df.iloc[2, 3]
```

```python
# 0, 1행의 2, 3열 추출
df.iloc[[0, 1], [2, 3]]
```
